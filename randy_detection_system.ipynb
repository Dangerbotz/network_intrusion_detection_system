```python
normalPacketPercent = float(len(df[df.label==0]))/len(df)*100.0
attackPacketPercent = float(len(df[df.label==1]))/len(df)*100.0

print ('normal packet percentage: ',  normalPacketPercent)
print ('attack packet percentage: ',  attackPacketPercent)
```

```python
# now lets group with the new variable
df_grouped = df.groupby(by=['attack_cat','service'])
print ('Percentage of attacks in each group:\n')
print (df_grouped.label.sum() / df_grouped.label.count() * 100)
```

```python
# this python magics will allow plot to be embedded into the notebook
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore', DeprecationWarning)
%matplotlib inline
```

```python
# Start by just plotting what we previsously grouped!
plt.style.use('ggplot')

df_grouped = df.groupby(by=['attack_cat'])
attack_rate = df_grouped.label.sum() / df_grouped.label.count()
ax = attack_rate.plot(kind='barh')
```

```python
# the cross tab operator provides an easy way to get these numbers
normal = pd.crosstab([ df['attack_cat'] ], 
                       df.label.astype(bool))
print (normal)
```

```python
normal_rate = normal.div(normal.sum(1).astype(float),
                             axis=0) # normalize the value

# print normal non-attack rate
normal_rate.plot(kind='barh', 
                   stacked=True)
```

```python
# cross tabulate example from http://nbviewer.ipython.org/gist/fonnesbeck/5850463# 
normal_counts = pd.crosstab([df['attack_cat'], df['is_sm_ips_ports']],
                             df.label.astype(bool))
                             
normal_counts.plot(kind='bar', stacked=True)

# divide the counts to get rates
normal_rate = normal_counts.div(normal_counts.sum(1).astype(float),axis=0)
normal_rate.plot(kind='barh', stacked=True)
```
