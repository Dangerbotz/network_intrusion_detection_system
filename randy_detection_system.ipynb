```python
# the cross tab operator provides an easy way to get these numbers
attack = pd.crosstab([ df['attack_cat'] ], 
                       df.label.astype(bool))
print (attack)
```

```python
#attack_rate = attack.div(attack.sum(1).astype(float), axis=0) # normalize the value
attack_rate = attack.div(len(df.label), axis=0) # normalize the value
print ('attack_rate: \n', attack_rate)
# print normal non-attack rate
attack_rate.plot(kind='barh', stacked=True)

#len(df[df.readmitted != 0])
# print ( 'number of records: ', len(df.label) )  82332
```

```python
# cross tabulate example from http://nbviewer.ipython.org/gist/fonnesbeck/5850463# 
normal_counts = pd.crosstab( df['attack_cat'],
                             df.label.astype(bool))
                             
normal_counts.plot(kind='bar', stacked=True) 

# divide the counts to get rates
normal_rate = normal_counts.div(len(df.label),axis=0)
normal_rate.plot(kind='barh', stacked=True)
```

```python
import seaborn as sns
cmap = sns.diverging_palette(220, 10, as_cmap=True) # one of the many color mappings

df1 = df[['label', 'proto', 'state', 'dur', 'sbytes', 'dbytes', 'sttl', 'dttl', 'sloss', 'dloss']]
df2 = df[['label', 'service', 'sload', 'dload', 'spkts', 'dpkts', 'swin', 'dwin', 'stcpb', 'dtcpb']]
df3 = df[['label', 'smean', 'dmean', 'trans_depth', 'response_body_len', 'sjit', 'djit', 'tcprtt']]
```
